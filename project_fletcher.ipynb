```python
from pymongo import MongoClient
client = MongoClient()
db = client.yelp
```

```python
list(db.reviews.find().limit(5))
```

```python
list(db.business.find().limit(10))
```

```python
list(db.business.aggregate([
    {"$match": {'categories': {'$regex': 'Bar'}}},
    {"$match": {"review_count": {"$gte": 50}}},
    {"$count": "name"}]))
```

```python
list(db.business.aggregate([
    {"$match": {'categories': {'$regex': 'Bar'}}},
    {"$match": {"review_count": {"$gte": 50}}},
    {"$distinct": 'state'}]))
```
